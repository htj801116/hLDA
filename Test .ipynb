```python
from hlda import hLDA, draw_graph
```

```python
import matplotlib.pyplot as plt
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim

tokenizer = RegexpTokenizer(r'\w+')

# create English stop words list
en_stop = get_stop_words('en')

# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()

# create sample documents
doc_a = "US President Donald Trump said a major conflict with North Korea is possible in the standoff over its nuclear and missile programs."
doc_b = "The move by the Macron campaign comes as his campaign this week revealed it has been the target of at least five cyber-attacks since January."
doc_c = "The United States bolsters the existence of a justice system that is applied to other countries in the world and has been a leading contributor to the architecture of an international accountability system."
doc_d = "We call on all governments involved to refrain from using belligerent rhetoric and to strive for peaceful constructive dialogue."
doc_e = "He has championed a nuclear deal that ended a stand-off with the West â€“ and led to many sanctions being lifted in return for Tehran curbing its nuclear programme."

# compile sample documents into a list
doc_set = [doc_a, doc_b, doc_c, doc_d, doc_e]
# list for tokenized documents in loop
texts = []

for i in doc_set:
    # clean and tokenize document string
    raw = i.lower()
    tokens = tokenizer.tokenize(raw)
    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in en_stop]
    # stem tokens
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    # add tokens to list
    texts.append(stemmed_tokens)
```

```python
h = hLDA(texts, 10, 0.01, 0.5, 1, 1000, 3)
```

```python
graph = draw_graph(h)
```
